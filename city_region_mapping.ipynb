```python
geo_reviews = geo_reviews.drop(index=(geo_reviews[geo_reviews['city'].isna()].index))
```

```python
geo_reviews_cities_unique = geo_reviews['city'].unique()
geo_reviews_cities_unique[:10]
```

```python
from opencage.geocoder import OpenCageGeocode

def get_region_by_city(city_name):
    API_KEY = '7b17a0f59db9480fa2dbba87a17af66e'
    geocoder = OpenCageGeocode(API_KEY)
    
    result = geocoder.geocode(city_name)
    
    if result and len(result):
        components = result[0]['components']
        return components.get('state', 'Регион не найден')
    return 'Город не найден'
```

```python
city_region_mapping = dict()
```

```python
for city in geo_reviews_cities_unique:
    if city not in city_region_mapping:
        city_region_mapping[city] = get_region_by_city(city)
```
