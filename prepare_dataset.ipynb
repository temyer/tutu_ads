```python
with open('city_region_mapping.json', 'r') as file:
    city_region_mapping = json.load(file)
```

```python
def read_geo_reviews():
    # https://github.com/yandex/geo-reviews-dataset-2023

    geo_reviews = pd.read_table('./geo-reviews-dataset-2023.tskv', header=None).dropna()
    columns = []

    for el in geo_reviews.iloc[0]:
        columns.append(el.split('=')[0])

    geo_reviews.columns = columns

    for col_name in geo_reviews.columns:
        geo_reviews[col_name] = geo_reviews[col_name].apply(lambda x: x.split('=')[1])

    geo_reviews['rating'] = geo_reviews['rating'].astype(float)

    geo_reviews['city'] = geo_reviews['address'].apply(lambda x: x.split(',')[0])
    geo_reviews.drop(columns=['address'], inplace=True, errors='ignore')
    geo_reviews['region'] = geo_reviews['city'].apply(lambda x: city_region_mapping.get(x, None))

    geo_reviews['rubrics_0'] = geo_reviews['rubrics'].apply(lambda x: x.split(';')[0])

    rich_hotel_feat = ['Гостиница', 'База, дом отдыха', 'Турбаза', 'Санаторий', 'Отдых на ферме',
              'Глэмпинг']

    not_rich_hotel_feat = [
        'Хостел'
    ]

    tour_feat = [
        'Экскурсии', 'Водные прогулки', 'Горнолыжная школа', 'Достопримечательность',
        'Музей', 'Памятник, мемориал', 'Жанровая скульптура', 'Памятник технике',
        'Могилы известных людей', 'Мемориальная доска, закладной камень',
        'Заповедник',
    ]

    nature_feat = [
        'Пляж', 'Лесной массив', 'Термальный источник', 'Горная вершина',
        'Водопад', 'Природа', 'Парк культуры и отдыха'
    ]

    all_feats = rich_hotel_feat + not_rich_hotel_feat + tour_feat + nature_feat


    def feat_adder(x):
        if x in (rich_hotel_feat):
            return 'rich_hotel_feat'
        if x in (not_rich_hotel_feat):
            return 'not_rich_hotel_feat'
        if x in (tour_feat):
            return 'tour_feat'
        if x in (nature_feat):
            return 'nature_feat'
        return 'other'

    geo_reviews_filtered = geo_reviews[geo_reviews['rubrics_0'].isin(all_feats)]
    geo_reviews_filtered['rubrics_common'] = geo_reviews_filtered['rubrics_0'].apply(feat_adder)

    geo_reviews_ratings = geo_reviews_filtered.groupby(['region', 'rubrics_common'], as_index=False).agg(
        rating_mean = ('rating', 'mean'),
        rating_std = ('rating', 'std')
    )

    geo_reviews_grp = geo_reviews_filtered.groupby('region', as_index=False).agg(
        rich_hotel_count = ('rubrics_0', lambda r: r.isin(rich_hotel_feat).sum()),
        not_rich_hotel_count = ('rubrics_0', lambda r: r.isin(not_rich_hotel_feat).sum()),
        tour_count = ('rubrics_0', lambda r: r.isin(tour_feat).sum()),
        nature_feat = ('rubrics_0', lambda r: r.isin(nature_feat).sum()),
    )
    
    tmp = pd.pivot_table(geo_reviews_ratings, values=['rating_mean', 'rating_std'], index='region',
                       columns=['rubrics_common'])

    tmp.columns = [f"{i[1]}_{i[0]}" for i in tmp.columns]
    tmp.reset_index(inplace=True)

    geo_final = geo_reviews_grp.merge(tmp, on='region')

    geo_final = geo_final[~geo_final['region'].isin(['Moscow', 'Moscow Oblast', 'Moskovskaya', 'Saint Petersburg', 'Leningrad Oblast'])]

    geo_final['region'] = geo_final['region'].apply(lambda x: region_mapping_2[x.strip()])
    geo_final['region'] = geo_final['region'].apply(lambda x: region_mapping[x.strip()])

    return geo_final
```

```python
def read_tutu_data():
    to_cluster = pd.read_csv('to_cluster.csv')
    to_cluster.drop(columns=['Unnamed: 0'], inplace=True)
    
    return to_cluster
```

```python
def read_rosstat_data():
    region_feat = pd.read_csv('region_feat2.csv')

    return region_feat
```

```python
def read_full_data():
    to_cluster = read_tutu_data()
    geo_final_filtered = read_geo_reviews()
    region_feat = read_rosstat_data()

    full_data = to_cluster.merge(geo_final_filtered, on='region', how='left').merge(region_feat, on='region')

    return full_data
```
